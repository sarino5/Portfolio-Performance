```python
import pandas as pd
import yfinance as yf
import numpy as np
import random
import matplotlib.pyplot as plt
import statistics as stats
import seaborn as sns
```

```python
nyse_df = pd.read_csv("nasdaq_screener_1611715051396.csv")
nyse_df.rename(columns={'Symbol':'Ticker','Name':'Company'}, inplace = True)
nyse_df.drop(["Company","Last Sale", "Net Change", "% Change", "Market Cap", "Country", "IPO Year", "Volume", "Sector", "Industry"], axis = 1, inplace = True)
```

We see some symbols(. and /) in a Ticker, which makes the stock either non-existent or unrecognizable by the yfinance package. Those with '.' must be dropped, and the '/' must be converted into '-' to be recognizable by yfinance

```python
nyse_df = nyse_df.drop(nyse_df[(nyse_df.Ticker.str.contains("\\.")) | (nyse_df.Ticker.str.contains("SCA"))].index)                              
nyse_df["Ticker"] = nyse_df["Ticker"].str.replace("/", "-")
```

```python
fifteen_stocks = 15
```

```python
def portfolio(fifteen_stocks):
    
    portfolio_unchecked = []
    for index, rows in nyse_df.iterrows():
        portfolio_unchecked.append(rows.Ticker)
    portfolio_unchecked = random.sample(portfolio_unchecked, len(portfolio_unchecked))
    portfolio_unchecked = portfolio_unchecked[0:fifteen_stocks]
    
    portfolio = []
    for x in portfolio_unchecked:
        portfolio.append(x.strip())
    return portfolio

#print(portfolio(fifteen_stocks))
```

```python
def avg_stock_return(tickers,years):
    return_list = []
    for ticker in tickers:
        close = yf.download(ticker, period = str(years) + "y",interval='1d', group_by='Ticker').loc[:, 'Close']
        avg_return = ((close[-1]/close[0])**(1/years)-1)*100
        avg_return = round(avg_return, 2)
        return_list.append(avg_return)
    return round(stats.mean(return_list), 2)
#print(avg_stock_return(portfolio(fifteen_stocks), 10))
```

```python
list_fifteen_stock_portfs_avg = []

for x in list(range(10)):
    list_fifteen_stock_portfs_avg.append(avg_stock_return(portfolio(fifteen_stocks), 10))
    
```
